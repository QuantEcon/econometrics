## Introduction

Knowing the true mechanism $P$, while we are endowed with a decision problem $(\mathcal{H},L,\mathcal{A})$, naturally leads to the best in class action $a^{*}_{L, P, \mathcal{A}}$ as the optimal decision rule. Although in reality we do not know $P$, it is illuminating to consider this case as it is natural to require that at least with a sufficiently large (in fact, infinite) sample, the true mechanism can be revealed. The property that guarantees this scenario in our case is the ergodicity of the assumed statistical models.  

This argument suggests the best in class action, $a^{*}_{L, P, \mathcal{A}}$, as the optimal asymptotic (large sample) solution. In other words, from any sensible decision rule a minimal property to require is that as the sample size grows, the distribution of the decision rule converges to a degenerate distribution concentrated at the point $a^{*}_{L, P, \mathcal{A}}$. As an illustration, consider again the MLE estimator in the coin tossing problem for which the above property, as the figure below demonstrates, obviously holds.

```python
theta0 = .8

def mle_dist(n , theta0):
    y_grid = np.arange(n + 1)                       # sum of ones in a sample
    decision_rule = y_grid/n
    true_dist = stats.binom(n, theta0)
    
    return y_grid, decision_rule, true_dist

fig, ax = plt.subplots(1, 2, figsize = (16, 5))
n = 100
yg, dr, tr = mle_dist(n, theta0)
ax[0].plot(dr, tr.pmf(yg), 'o')
ax[0].set_xlim([.6, 1])
ax[0].vlines(dr, 0, tr.pmf(yg), lw = 5, alpha = .8, color = colors[0])
ax[0].axvline(theta0, color = colors[2], lw = 2, label = r'True $\theta_0$')

n = 10000
yg, dr, tr = mle_dist(n, theta0)
ax[1].plot(dr, tr.pmf(yg), 'o')
ax[1].set_xlim([.6, 1])
ax[1].vlines(dr, 0, tr.pmf(yg), lw = 5, alpha = .8, color = colors[0])
ax[1].axvline(theta0, color = colors[2], lw = 2, label = r'True $\theta_0$')
```

This property is called *consistency* and one objective of this notebook is to investigate the conditions under which it can be established.   

Given that asyptotically the true $P$ can be learned, a natural question to ask is why not to set $\mathcal{A}$ large enough so as to guarantee $\gamma(P)\in\mathcal{A}$, i.e. a correctly specified model. We will see the sense in which requiring consisteny also ties our hands in terms of the "size" of $\mathcal{A}$. 

## Consistency of decision rules

To start out our asymptotic inquiry, we first define consistency in a more precise manner than we did before. There are two, slightly different notions depending on the tractability and objectives of the decision problem at hand.  
- **Consistency in terms of the loss function:** a decision rule, $d: \mathcal{S} \mapsto \mathcal{A}$, is consistent in terms of the loss function relative to $(\mathcal{H}, \mathcal{A})$, if for all $P \in \mathcal{H}$ the following holds
\begin{align*}
	& L(P, d(Z^n)) \ \ \underset{n \to \infty}{\overset{\text{P}}{\rightarrow}} \ \ \inf_{a \in \mathcal{A}}L(P, a) = L\left(P, a^{*}_{L, P,\mathcal{A}}\right)
\end{align*}

- **Consistency in terms of the action:** a decision rule, $d: \mathcal{S} \mapsto \mathcal{A}$, is consistent in terms of the action relative to $(\mathcal{H}, \mathcal{A})$, if for all $P \in \mathcal{H}$ the following holds
\begin{align*}
P\left\{z^{\infty} \big| \lim_{n \to \infty} m\left(d(z_n), a^*_{L, P, \mathcal{A}}\right) > \epsilon\right\} = 0 \quad \text{for} \quad \forall\epsilon>0.
\end{align*}
The necessary condition for this notion is the identifyability of $a^*_{L, P, \mathcal{A}}$ that we define as follows:
> *Identifiability:* $a^*_{L, P, \mathcal{A}}$ is identified relative to $(P, \mathcal{A})$ if $a^*_{L, P, \mathcal{A}}$ is the unique minimizer of $L(P, \cdot)$ over $\mathcal{A}$.
    
Under identification, the two notions are equivalent. As the above definitions suggest, however, the former is more general to the extent that it also allows for partially identified statistical models. Unless otherwise noted, we will work with consistency in terms of the loss function and call it simply *consistency*.  

Note that in the above definitions, the set of assumed statistical models, $\mathcal{H}$ plays a key role: it outlines the set of distributions under which the particular notion of convergence is required. Ideally, we want the convergence to hold under the true distribution, however, because we do not know $P$, the "robust" approach is to be as agnostic as possible regarding $\mathcal{H}$. The central insight of statstical decision theory is to highlight that this approach is not limitless: one needs to find a balance between $\mathcal{H}$ and $\mathcal{A}$ to obtain decision rules with good properties.

Consistency has strong implications for the risk functional in large samples: the degenerate limiting distribution of the decision rule implies that asymptotically the variability of the decision rule vanishes and so 
$$R_{\infty}(P,d) \overset{a.s.}{=} L\left(P, a^{*}_{L, P,\mathcal{A}}\right)$$


##  Uniform law of large numbers


Although it is hard to provide generally applicable sufficient conditions for consistency, roughly speaking, we can identify two big classes of decision rules for which powerful results are available.
 * Bayes decision rules 
 * Frequentist estimators building on the empirical distribution $P_n$ (NOTE: somewhere we should define $P_n$)

In the following we will be focusing on the latter case. The basic idea behind this approach to finding consistent decision rules is to utilize the consistency definitions directly, but instead of using the quantity $L(P, a)$ that we cannot actually evaluate, subsitute the empirical distribution into the loss function and work with the empricial loss $L(P_n, a)$ instead. MANSKI: empirical distribution is not in the loss functions's domain

Given that the loss function is continuous in its first argument, the law of large numbers (LLN), accompanied with the continuous mapping theorem, implies

$$L(P_n, a) \quad  \underset{n \to \infty}{\to} \quad L(P, a)$$

that is, for a *fixed action* $a$, the emprical loss converges to the true loss as the sample size goes to infinity. Notice, however, that consistency 
 - is not a property of a given action, but a whole decision rule, more precisely, it is about the convergence of the sequence $\{d(z^n)\}_{n\geq 1} \subseteq \mathcal{A}$ 
 - requires convergence to the *minimized* loss (within $\mathcal{A}$) that we ought to take into account while considering possible decision rules 
 
Maybe the most intuitive way to proceed, in the spirit of the analogy principle, is to define decision rules by minimizing the *empirical loss*. 

$$d(z^n) := \arg\min_{a\in\mathcal{A}} L(P_n, a)$$

where the dependence on the sample is embodied by the empirical distribution $P_n$. The heuristic idea is that because $d(z^n)$ minimizes $L(P_n, a)$, which itself converges to $L(P, a)$, then $d(z^n)$ should go to $a^{*}_{L, P, \mathcal{A}}$, the minimizer of $L(P, a)$, as the sample size grows. 

What do we need to ensure this argument to hold? First, note that the standard law of large numbers is inadequate for this purpose:
> If $\{a_n\}_{n\geq 1}$ is a nonstochastic sequence in $\mathcal{A}$, the law of large numbers says that for every $a_m \in \{a_n\}_{n\geq 1}$, there exists an integer $N(a_m, \varepsilon)$, such that for all $n>N(a_m, \varepsilon)$, the probability of $\left| L(P_n, a_m) - L(P, a_m) \right| < \varepsilon$ converges to one. Because $N$ depends on $a_m$, it is possible that there is no $n$ so that $n>N(a_n, \varepsilon)$, hence $L(P_n, a_n)$ may never approach $L(P, a_n)$.

What we need is to require that for all $\varepsilon>0$, there exists an integer $N(\varepsilon)$ *independent of $a$*, such that for all sample sizes larger than $N$, 

$$\lim_{n\to \infty} P\left\{\sup_{a \in\mathcal{A}} \left| L(P_n, a) - L(P, a) \right| < \varepsilon\right\}= 1$$

This notion is called uniform law of large numbers referring to the fact that the convergence is guaranteed simultaneously for all functions in $\mathcal{A}$. 

As an illustration ...

```python
theta0 = .7
n = 10000

true_dist = stats.binom(1, theta0)

def relative_entropy(theta0, a):
    return theta0 * np.log(theta0/a) + (1 - theta0) * np.log((1 - theta0)/(1 - a))

def empirical_loss(prob_0, a):    
    return prob_0 * np.log(prob_0/a) + (1 - prob_0) * np.log((1 - prob_0)/(1 - a))

a_grid = np.linspace(0.1, .99, 100)               # action space represented as [0, 1]
loss = relative_entropy(theta0, a_grid)            # form of the loss function

sample = np.asarray([1, 1, 0, 1, 0, 0, 0, 1, 0, 1])
sample = np.hstack([sample, true_dist.rvs(n)])

sample_sizes = [10, 50, 100, 500, 1000, 10000]
mle_est = [sample[:i].mean() for i in sample_sizes]
emp_loss = [empirical_loss(i, a_grid) for i in mle_est]


# ================================
# Plot
# ================================

fig, ax = plt.subplots(2, 2, figsize = (16, 10))
ax[0, 0].set_title('Figrue', fontsize = 17)
ax[0, 0].plot(a_grid, loss, lw = 7, label = r'Loss')
ax[0, 0].axhline(0, color = 'k')
ax[0, 0].axvline(theta0, lw = 2, color = colors[2])

ax[0, 0].plot(a_grid, emp_loss[0], color = colors[1], lw = 3, alpha = .5, label = r'Emprical Loss $(n=5)$')
ax[0, 0].plot(mle_est[0], empirical_loss(mle_est[0], mle_est[0]), 'o')
ax[0, 0].plot(mle_est[0], empirical_loss(mle_est[1], mle_est[0]), 'o')
ax[0, 0].plot(mle_est[0], empirical_loss(mle_est[2], mle_est[0]), 'o')
ax[0, 0].plot(mle_est[0], empirical_loss(mle_est[3], mle_est[0]), 'o')
ax[0, 0].axvline(mle_est[0], ymin = 0.0, ymax = 0.14, lw = 3, alpha = .6, linestyle = '--', color = colors[1])

ax[0, 0].plot(a_grid, emp_loss[1], color = colors[2], lw = 3, alpha = .5, label = r'Emprical Loss $(n=10)$')
ax[0, 0].axvline(mle_est[1], ymin = 0.0, ymax = 0.14, lw = 2, alpha = .6, linestyle = '--', color = colors[2])

ax[0, 0].legend(loc = 'best', fontsize = 15)

ax[0, 0].set_ylim([-.03, .2])
ax[0, 0].set_xlim([.4, .85])

np.random.seed(19871002)
sample_sizes = np.arange(10, 800)
sample = true_dist.rvs(n)

ax[0, 1].set_title('Difference between the empricial and true loss for given $a$', fontsize = 15)
ax[0, 1].axhline(0, color = 'k')
emp_loss = [empirical_loss(sample[:i].mean(), .2) for i in sample_sizes]
ax[0, 1].plot(sample_sizes, emp_loss - relative_entropy(theta0, .2), label = r'$a=.2$')
emp_loss = [empirical_loss(sample[:i].mean(), .9) for i in sample_sizes]
ax[0, 1].plot(sample_sizes, emp_loss - relative_entropy(theta0, .9), label = r'$a=.9$')
emp_loss = [empirical_loss(sample[:i].mean(), .7) for i in sample_sizes]
ax[0, 1].plot(sample_sizes, emp_loss - relative_entropy(theta0, .7), label = r'$a=.7$')
ax[0, 1].legend(loc = 'best', fontsize = 16)
ax[0, 1].set_xlabel(r'Sample size $(n)$', fontsize = 13)


emp_loss = [empirical_loss(sample[:i].mean(), a_grid) for i in sample_sizes]
ax[1, 0].axhline(0, color = 'k')
ax[1, 0].plot(a_grid, emp_loss[0]-loss, color = colors[0], lw = 3, alpha = .5, label = r'$n=10$')
ax[1, 0].plot(a_grid, emp_loss[10]-loss, color = colors[1], lw = 3, alpha = .5, label = r'$n=20$')
ax[1, 0].plot(a_grid, emp_loss[30]-loss, color = colors[2], lw = 3, alpha = .5, label = r'$n=40$')
ax[1, 0].plot(a_grid, emp_loss[50]-loss, color = colors[3], lw = 3, alpha = .5, label = r'$n=60$')
ax[1, 0].plot(a_grid, emp_loss[80]-loss, color = colors[4], lw = 3, alpha = .5, label = r'$n=90$')
ax[1, 0].set_title(r'Excess loss (empirical loss - true loss)')
ax[1, 0].set_xlabel(r'Action space, $a$')
ax[1, 0].legend(loc = 'best', fontsize = 14)

```

### Special case (plug-in estimators)

There are special cases where we do not need to worry about uniformity. When the true loss function has a minimizer $a^{*}_{L, P, \mathcal{A}}$ that admits a closed form in the sense that it can be expressed as a continuous function of $P$, the empirical loss minimizer is simply a sample analog of $a^{*}_{L, P, \mathcal{A}}$. Effectively, there is no need to minimize the emprical loss, so in spirit, it is as if we kept a particular action fixed. In these cases the standard LLN is sufficient to establish consistency. 
 
   * *Sample mean:* 
       > Consider the quadratic loss function with the feature $\gamma(P) = E[Z]$, i.e. the loss is $L(P,a) = E[(a - E[Z])^2]$. Evidently, the minimizer of this quantity is $E[Z]$. Although we could proceed by minimizing the empricial loss to derive the decision rule, we do not need to do that, because we know up front that it is equal to the sample analog of $E[Z]$. The decision rule is the plug-in estimator $d(z^n) = \frac{1}{n}\sum_{i=1}^{n}z_i$.
   * *OLS estimator:*
       > Suppose that $Z=(Y, X)$ and consider the quadratic loss $L(P, a) = E[(Y-a)^2 | X]$, where $E[\cdot | X]$ is the expectation operator conditioned on $X$. Assume that $(Y, X)$ follows a multivariate normal distribution, then the minimizer of $L$ is given by $a = E\left[(X'X)^{-1}\right]E[X'Y]$. Consequently, there is no need for explicit minimization, we can use the least squares estimator as a plug-in estimator for $a$.  

## Uniform law of large numbers and consistency

In more general nonlinear models, however, $a^{*}_{L, P, \mathcal{A}}$ has no analytical form, so we cannot avoid minimizing the empirical loss in order to derive the decision rule. In these cases, guaranteeing the validity of uniform convergence becomes essential. In fact, it can be shown that consistency of any decision rule that we derive by minimizing some empirical loss function, is *equivalent with* the uniform LLN. 

>**Vapnik and Chernovenkis (1971)**: Uniform convergence 
$$\lim_{n\to \infty} P\left\{\sup_{a \in\mathcal{A}} \left| L(P_n, a) - L(P, a) \right| < \varepsilon\right\}= 1$$
for all $\varepsilon >0$ is *necessary and sufficient* for consistency of decision rules arising from empirical loss minimization over $\mathcal{A}$. 


Although this characterization of consistency is theoretically intriguing, it is not all that useful in practice unless we find easy-to-check conditions which ensure uniform convergence of empirical loss over a particular function class $\mathcal{A}$. As a preparation to discuss these conditions, introduce first some notation

Almost all loss functions used in practive can be cast in the structure taking the following form. There exist 
 - $l : \mathcal{A}\times Z \mapsto R^{m}$ and 
 - a *continuous* $r : R^{m} \mapsto R_+$ such that
    
    $$ L(P, a) = r\left( \int_Z l(a, z)dP (z)\right) $$

Continuity of $r$ implies that in order to establish consistency it is enough to investigate the properties of the limit 

$$P\left( \sup_{a\in\mathcal{A}} \left| \int_Z l(a, z)dP_n (z) - \int_Z l(a, z)dP(z)\right| > \varepsilon \right) \quad  \underset{n \to \infty}{\to} \quad  0 \quad\quad \forall \varepsilon > 0$$ 

Formally, this requires that the class of functions $\left\{l(a, \cdot) : a \in \mathcal{A} \right\}$ is a **Glivenko-Cantelli class** for $P$.

> **Definition (Glivenko-Cantelli class):** Let $\mathfrak{F}$ be a class of integrable real-valued functions of the random variable $Z$ having distribution $P$. Consider the random variable 
$$\Vert P_n - P \Vert_{\mathfrak{F}} : = \sup_{f\in\mathfrak{F}} \left| \frac{1}{n}\sum_{i=1}^{n} f(z_i) - E[f] \right|$$
We say that $\mathfrak{F}$ is a Glivenko-Cantelli class for $P$, if $\Vert P_n - P \Vert_{\mathfrak{F}}$ converges to zero in probability as $n\to\infty$.

### Sufficient conditions

Variables of the form $\Vert P_n - P \Vert_{\mathfrak{F}}$ are ubiquitos in statistics and econometrics and there are well-known sufficient conditions that guarantee its convergence.

 * For parametric estimation, probably the most widely used (at least in econometrics) sufficient condition includes the following three assumptions (or some form of thereof). Suppose that the action space $\mathcal{A}$ is indexed by a finite dimensional vector $\theta\in\Theta \subset \mathbb{R}^{p}$. If
  1. $\Theta$ is *compact*
  2. $l(\theta, z)$ is a function such that $l(\cdot, z)$ is *continuous* on $\Theta$ with probability one
  3. $l(\theta, z)$ is *dominated* by a function $G(z)$, i.e. $|l(\theta, z)|\leq G(z)$ for all $\theta\in\Theta$, such that $E[G]<\infty$
   
then $\mathfrak{L} := \left\{l(a, \cdot) : a \in \mathcal{A} \right\}$ is a Glivenko-Cantelli class for $P$ and so the estimator that it represents is consistent. Among others, these assumptions are the bases for consistency of the (Quasi-)Maximum Likelihood (White, 1994) and the Generalized Method of Moments estimators (Hansen, 1982).  


* A somewhat different approach focuses on the "effective size" of the class $\left\{l(a, \cdot) : a \in \mathcal{A} \right\}$ and frames sufficient conditions in terms of particular complexity measures such as the Vapnik-Chernovenkis dimension or Rademacher complexity. The idea is to find tail bounds for the probability that $\Vert P_n - P \Vert_{\mathfrak{L}}$ deviates substantially above zero in terms of the complexity of $\mathfrak{L}$. In the following section we consider main principles in the derivation of non-asymptotic tail bounds.


### An interesting necessary condition

Only in the case of indicator functions under iid sampling, finiteness of the VC-dimension is a necessary and sufficient condition for two-sided uniform convergence. 


## Non-asymptotic bounds


**General outline and motivation**

* We looked at consistency of analog decision rules in terms of the unifrom law of large numbers.

* The uniform law of large numbers states that the sample average is uniformly converging to the expectation over a set of random variables.

* Consistency alone remains silent about the statistical properties of our estimator in finite samples. 

* Given a consistent analog estimator we would like to characterize the rate at which the convergence takes place.

* The rate of convergence will give us information regarding the size of the sample above which the asymptotic results are going to be a good approximation.

* In order to study the rate of convergence of analog decision rules we introduce concentration inequalities in finite samples and see how they get tighter as we increase the sample size.

* These statements are called concentration and tail bounds and they concern the region where most of the probability mass of a random variable is concentrated.

* First, we are going to see how the concentration measures work for a single random variable. This is going to parallel the law of large numbers.

* Second, we are going to see how the concentration measures work uniformly over a class of random variables. This is going to parallel the uniform law of large numbers.

* The key characteristic of concentration bounds which hold uniformly over classes of random variables is their complexity measure.


## Tail and Concentration Bounds

Throughout the section we should keep in mind the characterization of consistency in the case of analog estimators. This was equivalent to uniform convergence which we repeat here

$$P\left( \sup_{a\in\mathcal{A}} \left| \int_Z l(a, z)dP_n (z) - \int_Z l(a, z)dP(z)\right| > \varepsilon \right) \quad  \underset{n \to \infty}{\to} \quad  0 \quad\quad \forall \varepsilon > 0.$$

Using the developed notation denote the centered random variables for each action as

\begin{equation}
(P_n l_a - P l_a) := \int_Z l(a, z)dP_n (z) - \int_Z l(a, z)dP(z).
\end{equation}
It will be useful to study the tail probabilities of these centered random variables.

### Markov's inequality

Markov's inequality states that for any non-negative scalar random variable $Z$ and $t>0$ we have 

$$P\{Z \geq t\}\leq \frac{\mathbb{E}[Z]}{t}. $$

The idea behind this inequality is fairly simple and in fact it follows from the following obvious inequality 

$$ \mathbb{1}\{Z\geq t\}\leq \frac{Z}{t}$$

illustrated by the figure below. Clearly, for any probability measure $P$ over $Z$ with a finite first moment the Markov's inequality can be established.

```python
z = np.linspace(0, 10, 1000)
t = 3

f1 = (z >= t)
f2 = z / t

fig, ax = plt.subplots(1, 2, figsize = (16, 5))
#ax[0].arrow(0, 0, 10, 0, head_width=0.1, head_length=0.5, fc='k', ec='k')
#ax[0].arrow(0, 0, 0.5, 0.5, head_width=0.05, head_length=0.1, fc='k', ec='k')

ax[0].add_artist(ConnectionPatch(xyA=(0, 0), xyB=(10, 0), coordsA="data", coordsB="data",
                                 arrowstyle="->", mutation_scale=10, lw = 2))
ax[0].add_artist(ConnectionPatch(xyA=(0, 0), xyB=(0, 4), coordsA="data", coordsB="data",
                                 arrowstyle="->", mutation_scale=10, lw = 2))

ax[0].plot(z, f1, color = colors[0], lw = 3, label = r'$\mathbf{1}\{Z \geq t\}$')
ax[0].plot(z, f2, color = colors[1], lw = 3, label = r'$Z/t$')
#ax.legend(bbox_to_anchor=(1.05, 1), loc=2, fontsize=14)
ax[0].set_xticklabels([])
ax[0].set_yticklabels([])
ax[0].text(9.3, -.25, r'$Z$', fontsize = 17)
ax[0].text(8, 3, '$Z/t$', fontsize = 16)
ax[0].text(5, .7, '$\mathbf{1}\{Z \geq t\}$', fontsize = 16)
ax[0].text(t - .05, -.25, '$t$', fontsize = 16)
ax[0].text(0, -.25, '$0$', fontsize = 16)
ax[0].axhline(1, linestyle = '--', color = 'k', lw = 1)
ax[0].set_title('Logic behind Markov\'s inequality', fontsize=18)
ax[0].set_ylim([0, 4])

t = 2.5
mu = 2
z = np.linspace(mu - 6, mu + 6, 1000)

g1 = (z >= mu + t) + (z <= mu - t)
g2 = ((z - mu) / t)**2

ax[1].add_artist(ConnectionPatch(xyA=(mu - 6, 0), xyB=(mu + 6, 0), coordsA="data", coordsB="data",
                                 arrowstyle="->", mutation_scale=10, lw = 2))
ax[1].add_artist(ConnectionPatch(xyA=(mu - 6, 0), xyB=(mu - 6, 4), coordsA="data", coordsB="data",
                                 arrowstyle="->", mutation_scale=10, lw = 2))
ax[1].plot(z, g1, color = colors[0], lw = 3, label = r'$\mathbf{1}\{Z \geq t\}$')
ax[1].plot(z, g2, color = colors[1], lw = 3, label = r'$Z/t$')
ax[1].set_xticklabels([])
ax[1].set_yticklabels([])
ax[1].text(mu + 5.3, -.25, r'$Z$', fontsize = 17)
ax[1].text(4.7, 3, r'$\frac{|Z-\mu|^2}{t^2}$', fontsize = 20)
ax[1].text(4.7, .7, r'$\mathbf{1}\{|Z -\mu |\geq t\}$', fontsize = 16)
ax[1].text(mu - .05, -.25, r'$\mu$', fontsize = 16)
ax[1].text(mu + t - .15, -.25, r'$\mu + t$', fontsize = 16)
ax[1].text(mu - t - .15, -.25, r'$\mu - t$', fontsize = 16)
#ax[1].axhline(1, linestyle = '--', color = 'k', lw = 1)
ax[1].set_title('Logic behind Chebyshev\'s inequality', fontsize=18)
ax[1].set_ylim([0, 4])
plt.show()
```

It follows from Markov's inequality that for any strictly monotonically increasing non-negative function $\phi$ and any random variable $Z$ (not necessarily non-negative) we have that

$$P\{Z \geq t\} = P\{\phi(Z) \geq \phi(t)\} \leq \frac{\mathbb{E}[\phi(Z)]}{\phi(t)}.$$

Taking the random variable $| Z - \mathbb{E}[Z]|$ and $\phi(x) = x^q$ for some $q>0$ leads us to tail bounds expressed in terms of the moments of the random variable $Z$.

$$P\{| Z - \mathbb{E}[Z]| \geq t\} \leq \frac{\mathbb{E}[| Z - \mathbb{E}[Z]|^q]}{t^q}.$$

Note that for $q = 2$, this form delivers the Chebyshev inequality (see the right panel of the Figure). This approach to bounding tail probabilities is quite general. Controlling higher order moments leads to (weakly) tighter bounds on the tail probabilities.

### Chernoff bounds

A related idea is at the core of the so called Chernoff bounds. There one takes the transformation $\phi(x) = e^{\lambda x}$ for the centered random variable $(Z - \mathbb{E}[Z])$ yielding

$$P\{(Z - \mathbb{E}[Z]) \geq t\} \leq \frac{\mathbb{E}\left[e^{\lambda(Z - \mathbb{E}[Z])}\right]}{e^{\lambda t}}.$$

Minimizing the bound over $\lambda$ (provided the moments exist) would lead us to the Chernoff bounds. We are close to apply the developed tail bounds to the random variable defined above -- $(P_n l_a - P l_a)$. Before however, it is going to be useful to exploit the fact that it is a sum of random variables.

### Hoeffding bounds

Suppose first that the sample $z^n$ is generated by an iid process. We are going to bound the tails of $(P_n l_a - P l_a)$ by using Chernoff's method.

First, denote the random variable $l(a, z_i) = z^a_i$ and rewrite $(P_n l_a - P l_a) = \frac{1}{n}\sum_{i=1}^n z^a_i - \mathbb{E}[z^a]$.

Chernoff's method would lead us to 

$$P\left\{\left(\frac{1}{n}\sum_{i=1}^n z^a_i - \mathbb{E}[z^a]\right) \geq t\right\} \leq \frac{\mathbb{E}\left[e^{\lambda \frac{1}{n}\sum_{i=1}^n \left(z^a_i - \mathbb{E}[z^a]\right)}\right]}{e^{\lambda t}} = e^{-\lambda t} \prod_{i=1}^n \mathbb{E}\left[e^{\lambda \frac{1}{n}\left(z^a_i - \mathbb{E}[z^a] \right)} \right]$$

where the equality follows from independece. Now, the problem of deriving a tight bound will boil down to deriving a tight bound on the moment generating function of $\left(z^a_i - \mathbb{E}[z^a] \right)$.

For our purposes a particularly important class of random variables will be the so called sub-Gaussian variables.

> **Definition.** A random variable $Z$ is called sub-Gaussian if there is a positive number $\sigma$ such that 
$$ \mathbb{E}\left[e^{\lambda \left(Z - \mathbb{E}[Z] \right)} \right] \leq e^{\sigma^2\lambda^2/2}$$ for all $\lambda \in \mathbb{R}$.

Remark: A Guassian variable with variance $\sigma^2$ is sub-Gaussian with parameter $\sigma$. There are other non-Guassian random variables which are sub-Gaussian -- bounded rvs, Rademacher rvs ...

> **Theorem.** (Hoeffding) Let the variables $z^a_i$ be iid and sub-Gaussian with parameter $\sigma$. Then, for all $t>0$ we have that
$$P\left\{\left|\frac{1}{n}\sum_{i=1}^n z^a_i - \mathbb{E}[z^a]\right| \geq t\right\} \leq 2\exp\left\{-\frac{t^2 n}{2\sigma^2}\right\}.$$

For a single action $a$ we see how the Hoeffding inequality is at the core of the law of large number. In fact it will give an exponential rate of convergence.

However, in order to talk about concentration properties of decision rules, $d: \mathcal{S}\mapsto\mathcal{A}$ we would like to make statements about tail probabilities uniformly over a class of actions. This leads us to uniform bounds.

### Uniform Tail and Concentration Bounds for classes of finite cardinality

As a first simple example we are going to consider the case when there are only finitely many actions in the set $\mathcal{A}$. Denote the cardinality by $\#\mathcal{A}=A_k$. A conservative estimate of the uniform tail bound in this case is going to be the union bound.

> **Corollary.** (Hoeffding union bound) Let the variables $\{z^a_i: i=1,\ldots, A_k\}$ be iid and sub-Gaussian with common parameter $\sigma$. Then, for all $t > 0$ we have that
$$P\left\{\sup_{a\in\mathcal{A}}\left| \frac{1}{n}\sum_{i=1}^n z^a_i - \mathbb{E}[z^a]\right| \geq t \right\} \leq \sum_{a = 1}^{A_k} P\left\{\left|\frac{1}{n}\sum_{i=1}^n z^a_i - \mathbb{E}[z^a]\right| \geq t \right\} \leq A_k 2\exp\left\{-\frac{t^2 n}{2\sigma^2}\right\}.$$

The difference between the uniform and individual Hoeffding bound is just a scaling factor, $A_k$. A clear difficulty we have to overcome is extending the above analysis to action spaces with infinite cardinality. For this end we need to find a better "measure" of the size of the action space. There are many such measures and usually the one which gives the sharpest bounds is context dependent.

### Uniform Concentration Bounds for classes of infinite cardinality

In order to work with sets of infinitely many functions we would like to capture the size of these infinite sets for the purpose of statistical analysis. One such measures which proves to be useful and have a nice intuitive interpretation is the Rademacher complexity of a function class.

For a given realization of a sample, $z^n$, of size $n$ consider the the set of vectors

$$ \mathcal{L}(z^n) := \left\{ (l(z_1), \ldots, l(z^n)) \in \mathbb{R}^n \mid l \in \mathcal{L} \right\}.$$

This is the number of ways we are able to label points of a sample using functions in the class $\mathcal{F}$. The empirical Rademacher complexity of $\mathcal{L}$ for fixed $z^n$ is defined as

$$ \mathcal{R}\left(\frac{\mathcal{L}(z^n)}{n} \right) := \mathbb{E}_{\epsilon}\left[\sup_{l\in \mathcal{L}}\Big| \frac{1}{n}\sum_{i=1}^n \epsilon_i l(z_i) \Big| \right] $$

where the expectation is taken with respect to the iid Rademacher random variables, $\epsilon_i$ which take value in $\{-1, 1\}$ with equal probability.

Then the Rademacher complexity of $\mathcal{L}$ is at sample size $n$

$$ \mathcal{R}_n\left(\mathcal{L}\right) := \mathbb{E}_{Z^n} \Big[\mathcal{R}\left(\frac{\mathcal{L}(z^n)}{n} \right)\Big] $$.

Note that the Rademacher complexity has an intuitive interpretation. It is the average of the maximum correlation between the vector $\big(l(z_1), \ldots, l(z_n)\big)$ and the pure noise vector $\big(\epsilon(z_1), \ldots, \epsilon(z_n)\big)$. The function class $\mathcal{L}$ is too "large" for statistical purposes if we can always choose a function, $l\in\mathcal{L}$ that has high correlation with a randomly drawn noise vector (Wainwright, 2015). 

### Detour: Illustrating Rademacher complexity

In the following we consider two classes of functions where the one with the higher number of free parameters has drastically smaller Rademcaher complexities. For better illustration we are going to consider the empirical Rademacher complexity, i.e. fixing a particular realization of the sample, $z^n$.

#### Linear model with intercept

Consider the case when $Z^n$ is iid sample according to the uniform distribution on $[0, 1]$. The class of functions is given by the affine class

$$\mathcal{L} := \{l : l(z) = \beta_0 + \beta_1 z, (\beta_0, \beta_1)\in \mathbb{R}^2 \}$$

Below we see how the maximum achievable correlation evolves as the sample size grows for a given realization of the sample and noise.

```python
n_slider = FloatSlider(min = 2, max = 50, step = 1, value = 10)

def d_OLS(Z, n):
        Y = Z[:, 0]
        X = np.stack((np.ones(int(n)), Z[:,1]), axis=-1)
        return np.linalg.inv(X.T @ X) @ X.T @ Y

def rademacher_OLS(n):
        '''Generates N Rademcaher random variables and a random sample.'''
        aux = np.random.rand(int(2*n))
        x = aux[::2]
        r = np.array([1 if x >= .5 else -1 for x in aux[1::2]])
        return x, r
    
def rademacher(n):
        '''Generates N Rademcaher random variables.'''
        aux = np.random.rand(int(n))
        return np.array([1 if x >= .5 else -1 for x in aux])

@interact(n = n_slider)
def rademacher_example(n):
    fig, ax = plt.subplots(figsize = (12, 8))

    np.random.seed(123)

    #x = np.linspace(.01, .99, int(n))
    #x = np.random.rand(int(n))
    x, y = rademacher_OLS(n)
    Z = np.stack((y, x), axis=-1)   

    [a, b] = d_OLS(Z, n)
    R = (a + b*x) @ y / int(n)

    aux = np.linspace(0, 1, 1000)

    ax.plot(aux, a + b*aux, color = colors[0])
    ax.scatter(x, y, color = 'r')
    ax.text(.8, 1.2, r'$\mathcal{R}' + r'_{}(x^{}, \epsilon^{}) = $'.format({int(n)}, {int(n)}, {int(n)}) + ' {:.4f}'.format(R), fontsize=18)
    ax.set_xlim([-.2, 1.2])
    ax.set_ylim([-1.5, 1.5])
    ax.set_title('Evolution of maximum attainable correlation\n for given sample and noise -- linear model', fontsize=18)
    plt.show()
```

Averaging over all potential realization of the noise vector would give the empirical Rademacher complexity.

```python
# number of simulations
sim = 1000

# max sample size
n_max = 100

R = []
for n in range(2, n_max, 1):
    x = np.linspace(.01, .99, int(n))

    R_n = []
    for i in range(sim):
        y = rademacher(n)
        Z = np.stack((y, x), axis=-1)
        [a, b] = d_OLS(Z, n)
        r = (a + b*x) @ y / int(n)
        R_n.append(r)
    R.append(np.array(R_n).mean())
```

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(R, color = colors[0], label=r'$\mathcal{R}_n(x^n)$')
ax.set_xlabel(r'$n$ -- sample size', fontsize=16)
ax.set_title('Empirical Rademacher convexity and sample size\nHints for rate of convergence', fontsize=18)
ax.legend(fontsize=18)
plt.show()
```

#### Sinusoid classification

Next, we consider a set of classifier functions where the classification boundary is given by a sine function.

$$\mathcal{L} := \Big\{ \mathbb{1}\{sin(az) \geq 0\} - \mathbb{1}\{sin(az) < 0\} : a \in \mathbb{R}_+ \Big\} $$

Remark: In order to have a closed form solution for the optimal classifier we are selecting the sample at convenient points. This is WLOG and useful for illustrative purposes.


```python
n_slider = FloatSlider(min = 2, max = 20 , step = 1, value = 5)

@interact(n = n_slider)
def rademacher_example(n):
    fig, ax = plt.subplots(figsize = (12, 8))

    x = np.array([2**(-(i+1)) for i in range(int(n))])
    y = rademacher(int(n))

    y_aux = np.array([1 if r > 0 else 0 for r in y])
    x_aux = np.array([2**((i+1)) for i in range(int(n))])

    a = np.pi*((1-y_aux)@x_aux + 1)

    c = np.array([1 if np.sin(a*x) > 0 else -1 for x in x])
    R = c @ y / int(n)


    x_axis = np.linspace(0, 1, 1000)

    ax.plot(x_axis, np.sin(a*x_axis), color = colors[0])
    ax.scatter(x, y, color = 'r')
    ax.set_xlim([-.2, 1.2])
    ax.set_ylim([-1.5, 1.5])
    ax.text(.8, 1.2, r'$\mathcal{R}' + r'_{}(x^{}, \epsilon^{}) = $'.format({int(n)}, {int(n)}, {int(n)}) + ' {:.4f}'.format(R), fontsize=18)
    ax.set_title('Evolution of maximum attainable correlation\n for given sample and noise', fontsize=18)
    plt.show()
```

As we can see, by choosing a sufficiently high frequency we can always find a curve which classifies the data perfectly. The Ramdemacher complexity alsways takes its maximum irrespective of the sample size. For statistical purposes the family of sine curves is too complex.

With the introduced concepts we are in the position to present an important concentration inequality for classes of infinite cardinality.

> **Theorem.**(see Bousquet et al. 2004) For a uniformly bounded set of functions, $\Vert l \Vert_{\infty}<B$ for all $ l\in\mathcal{L}$ and $\delta>0$ we have that

> $$ P\Big\{ \Vert P_n - P \Vert_{\mathcal{L}} \geq 2 \mathcal{R}_n(\mathcal{L}) + \delta \Big\} \leq 2 \exp\Big\{-\frac{n\delta^2}{8B^2} \Big\}. $$


Rewriting slightly makes it a bit clearer. If $\delta - 2 \mathcal{R}_n(\mathcal{L}) > 0$ then

> $$ P\Big\{ \Vert P_n - P \Vert_{\mathcal{L}} \geq  \delta \Big\} \leq 2 \exp\Big\{-\frac{n\big(\delta - 2 \mathcal{R}_n(\mathcal{L})\big)^2}{8B^2} \Big\}. $$

Hence, if $\mathcal{R}_n(\mathcal{L}) = o(1)$ then $\Vert P_n - P \Vert_{\mathcal{L}} \to 0$ or put it differently, $\mathcal{L}$ is Glivenko-Cantelli.

It is apparent from the above inequality that the tightness of the finite sample uniform bounds will be (partly) determined by the Rademacher complexity. The change in Rademacher complexity as the sample size grows will determine the rate of convergence.

Unfortunately, computing the Rademacher complexity directly is only feasible in rare special cases. There are vaious techniques however, which give bounds on the Rademacher complexity. For different classes of functions different techniques prove useful, so one usually proceeds on a case-by-case basis.

The most common ways of bounding the Rademacher complexity are via the Vapnik-Chervonenkis dimension for classification and via metric entropy for bounded real valued functions.
